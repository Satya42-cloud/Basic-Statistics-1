```python
sales2=sales.drop(['Date','Day','SKU','City','BU','Brand','Model'],axis=1)
sales2.head()
```

```python
def get_stand(x):
    return((x-x.mean())/x.std())
```

```python
stand_data=get_stand(sales2)
stand_data
```

```python
plt.hist(stand_data['Volume'])
```
